**Question 1: What is K-Nearest Neighbors (KNN) and how does it work in both
classification and regression problems?**

Answer:
K-Nearest Neighbors (KNN) is a non-parametric, instance-based machine learning algorithm used for both classification and regression tasks. It is called non-parametric because it makes no assumptions about the underlying data distribution. KNN is considered a lazy learning algorithm, meaning it does not explicitly train a model during the training phase but instead stores the training data and makes predictions only at the time of testing or query.

The core idea of KNN is to predict the output for a new data point based on the outputs of its ‘k’ closest neighbors in the feature space. The closeness or similarity between data points is typically measured using distance metrics such as Euclidean distance, Manhattan distance, or Minkowski distance.

How KNN works in Classification:

Given a new input sample, KNN calculates the distance between the sample and all points in the training dataset.

It identifies the k nearest neighbors based on the chosen distance metric.

The predicted class is determined using a majority voting rule among the k neighbors. That is, the class that appears most frequently among the neighbors is assigned to the new sample.

Optionally, weighted voting can be used where closer neighbors contribute more to the prediction than distant ones.

How KNN works in Regression:

The algorithm identifies the k closest neighbors to the new data point.

Instead of voting for a class, it averages the target values of the neighbors to predict the output.

Weighted averaging can also be used, where neighbors closer to the new point have higher influence on the predicted value.

Key Features of KNN:

Simple and intuitive, easy to implement.

Works well for small to medium-sized datasets.

Sensitive to the choice of k: a small k can lead to noisy predictions (overfitting), while a large k may smooth out local patterns (underfitting).

Requires careful feature scaling because distance metrics are sensitive to the magnitude of features.

Summary:
KNN predicts outcomes based on proximity to training samples rather than building an explicit model. In classification, it assigns the majority class among the nearest neighbors, while in regression, it predicts the average of neighbors’ target values. Its simplicity and effectiveness make it a popular choice for tasks like image recognition, recommendation systems, and medical diagnosis, though it can be computationally expensive for very large datasets.

**Question 2: What is the Curse of Dimensionality and how does it affect KNN
performance?**

Answer:
The Curse of Dimensionality refers to the various phenomena that arise when working with data in high-dimensional spaces (i.e., when the number of features is very large). As the dimensionality increases, the volume of the feature space grows exponentially, causing data points to become increasingly sparse. This sparsity creates challenges for machine learning algorithms that rely on distance or similarity measures, such as K-Nearest Neighbors (KNN).

How the Curse of Dimensionality affects KNN:

Distance Becomes Less Meaningful:
KNN relies on distance metrics (like Euclidean distance) to find the nearest neighbors. In high-dimensional spaces, the distances between data points tend to converge, meaning that the difference between the closest and farthest neighbors becomes negligible. As a result, KNN struggles to distinguish between nearby and distant points, reducing its predictive accuracy.

Increased Computational Complexity:
With more dimensions, the number of computations required to calculate distances for each query point increases dramatically. This leads to slower predictions, especially for large datasets, because KNN is a lazy learner that computes distances at prediction time.

Overfitting Risk:
High-dimensional data often contains many irrelevant or noisy features. KNN can give undue weight to these features, resulting in predictions that overfit the training data but fail to generalize to unseen data.

Requirement for More Data:
To maintain the same level of density in high-dimensional space, exponentially more training samples are required. Without sufficient data, KNN’s performance deteriorates because the nearest neighbors may not be truly representative of the local neighborhood.

Mitigation Strategies:

Dimensionality Reduction: Use techniques like Principal Component Analysis (PCA), t-SNE, or Feature Selection to reduce the number of features before applying KNN.

Feature Scaling: Standardize or normalize features to ensure that distances are meaningful.

Weighted Distance Metrics: Give more importance to relevant features while down-weighting irrelevant ones.

Summary:
The Curse of Dimensionality makes KNN less effective in high-dimensional spaces because distances lose meaning, computation becomes costly, and the algorithm is prone to overfitting. Careful feature selection, dimensionality reduction, and preprocessing are essential to maintain KNN’s performance when dealing with datasets with many features.

**Question 3: What is Principal Component Analysis (PCA)? How is it different from feature selection?**

Answer:
Principal Component Analysis (PCA) is a dimensionality reduction technique used in machine learning and statistics to reduce the number of features in a dataset while retaining as much of the original variance as possible. It transforms the original correlated features into a new set of uncorrelated variables called principal components (PCs). Each principal component is a linear combination of the original features, and they are ordered such that the first few components capture most of the variation in the data.

How PCA works:

Standardization: PCA typically begins with standardizing the features so that each has a mean of zero and a standard deviation of one. This ensures that features with larger scales do not dominate the analysis.

Covariance Matrix Calculation: Compute the covariance matrix of the standardized data to understand the relationships between features.

Eigen Decomposition: Calculate the eigenvalues and eigenvectors of the covariance matrix. Eigenvectors determine the direction of the principal components, and eigenvalues indicate the amount of variance explained by each component.

Projection: The original data is projected onto the top k eigenvectors (principal components), reducing dimensionality while retaining most of the original information.

Key Features of PCA:

Reduces dimensionality by creating new features (principal components) rather than selecting a subset of existing features.

Helps in visualization of high-dimensional data in 2D or 3D.

Can improve computational efficiency and reduce the risk of overfitting.

Difference Between PCA and Feature Selection:

Aspect	PCA	Feature Selection
Approach	Transforms original features into new components (linear combinations).	Selects a subset of original features without creating new ones.
Output	Principal components (new features).	Original features retained or removed.
Goal	Reduce dimensionality while preserving variance.	Reduce dimensionality while keeping the most relevant original features.
Interpretability	Less interpretable because PCs are combinations of multiple features.	More interpretable since selected features retain their original meaning.
Use Case	When correlations exist among features or for visualization.	When model interpretability is important or for removing irrelevant/noisy features.

Summary:
PCA is a dimensionality reduction technique that creates new, uncorrelated features to capture the maximum variance in the data, while feature selection identifies and keeps only the most relevant original features. PCA is ideal for reducing complexity in high-dimensional datasets, whereas feature selection is preferred when preserving the original meaning of features is important. Both techniques help improve model performance, reduce overfitting, and manage high-dimensional data effectively.

**Question 4: What are eigenvalues and eigenvectors in PCA, and why are they
important?**

Answer:
In Principal Component Analysis (PCA), eigenvalues and eigenvectors are fundamental mathematical concepts used to identify the directions of maximum variance in a dataset and to perform dimensionality reduction effectively.

1. Eigenvectors in PCA:
An eigenvector is a vector that indicates a direction in the feature space along which the data varies. In PCA, eigenvectors of the covariance matrix define the directions of the principal components. Each eigenvector represents a new axis in the transformed feature space, and the data is projected onto these axes to create the principal components.

2. Eigenvalues in PCA:
An eigenvalue is a scalar that represents the magnitude of variance along its corresponding eigenvector. In PCA, the eigenvalue associated with each eigenvector indicates how much of the total variance in the dataset is captured by that principal component. Larger eigenvalues correspond to directions with greater data variability, while smaller eigenvalues capture less important variation (often noise).

Importance of Eigenvalues and Eigenvectors in PCA:

Identifying Principal Components: Eigenvectors determine the directions of the new axes (principal components), while eigenvalues help rank these components by their importance.

Dimensionality Reduction: By selecting the top k eigenvectors with the largest eigenvalues, PCA reduces the number of features while preserving the majority of the variance in the data.

Variance Preservation: Eigenvalues quantify the proportion of variance retained in each principal component, guiding the choice of how many components to keep for effective data representation.

Noise Reduction: Components with small eigenvalues often represent noise rather than meaningful patterns. Discarding them improves model performance and simplifies analysis.

Example in PCA Workflow:

Compute the covariance matrix of standardized data.

Perform eigen decomposition to get eigenvalues and eigenvectors.

Sort eigenvectors by eigenvalues in descending order.

Project the data onto the top eigenvectors to obtain a reduced representation with maximum variance preserved.

Summary:
Eigenvectors in PCA define the directions of the new feature space (principal components), and eigenvalues quantify how much variance each component captures. Together, they allow PCA to reduce dimensionality effectively, preserve important patterns, and filter out noise, making them essential for visualization, feature reduction, and improving machine learning model performance.


**Question 5: How do KNN and PCA complement each other when applied in a single
pipeline?**

Answer:
K-Nearest Neighbors (KNN) and Principal Component Analysis (PCA) are two powerful machine learning techniques that, when combined in a pipeline, can improve model performance and efficiency, particularly for high-dimensional datasets. Each addresses a specific challenge: KNN is sensitive to the number of features and their scales, while PCA reduces dimensionality and captures the most relevant variance in the data.

How they complement each other:

Dimensionality Reduction Before KNN:

KNN relies on distance metrics (e.g., Euclidean or Manhattan) to find neighbors.

In high-dimensional datasets, the Curse of Dimensionality makes distance metrics less meaningful, reducing KNN’s accuracy.

PCA reduces the number of features by transforming the data into a set of principal components that capture the majority of variance, removing redundant or noisy features.

This ensures that KNN uses only the most informative dimensions, improving the reliability of neighbor distances.

Noise Reduction and Better Generalization:

PCA filters out less important dimensions, which often contain noise.

By removing noise, the distances computed by KNN are less affected by irrelevant features, leading to more robust predictions and better generalization on unseen data.

Computational Efficiency:

High-dimensional datasets increase the computational cost of KNN, as distances must be calculated for all features.

PCA reduces the number of features, making distance computations faster without significant loss of information, which is particularly useful for large datasets.

Visualization and Interpretability:

PCA can reduce features to 2 or 3 principal components, enabling visualization of data clusters.

Applying KNN after PCA allows intuitive understanding of neighbor relationships and decision boundaries in a lower-dimensional space.

Practical Example of a KNN + PCA Pipeline:

Standardize the dataset.

Apply PCA to reduce dimensionality while retaining, for example, 95% of variance.

Train a KNN classifier on the reduced dataset.

Make predictions on test data using the KNN classifier.

Summary:
PCA and KNN complement each other because PCA prepares the data by reducing dimensionality and removing noise, which addresses KNN’s sensitivity to high-dimensional spaces and irrelevant features. When combined, they lead to more accurate, efficient, and interpretable models, making this pairing particularly effective for datasets with many features or correlated variables.

**Question 6: Train a KNN Classifier on the Wine dataset with and without feature scaling. Compare model accuracy in both cases.**
(Include your Python code and output in the code box below.)

Answer:


```python
# Step 1: Import necessary libraries
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Step 2: Load the Wine dataset
data = load_wine()
X = data.data
y = data.target

# Step 3: Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# -------------------------------
# Case 1: KNN without feature scaling
# -------------------------------
knn_no_scaling = KNeighborsClassifier(n_neighbors=5)
knn_no_scaling.fit(X_train, y_train)
y_pred_no_scaling = knn_no_scaling.predict(X_test)
accuracy_no_scaling = accuracy_score(y_test, y_pred_no_scaling)
print("Accuracy without feature scaling:", accuracy_no_scaling)

# -------------------------------
# Case 2: KNN with feature scaling
# -------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = knn_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
print("Accuracy with feature scaling:", accuracy_scaled)
```

**Question 7: Train a PCA model on the Wine dataset and print the explained variance ratio of each principal component.**

```python
# Step 1: Import necessary libraries
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd

# Step 2: Load the Wine dataset
data = load_wine()
X = data.data
y = data.target

# Step 3: Standardize the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Step 5: Print the explained variance ratio for each principal component
explained_variance = pca.explained_variance_ratio_

# Display the variance ratio in a readable table format
pca_results = pd.DataFrame({
    'Principal Component': [f'PC{i+1}' for i in range(len(explained_variance))],
    'Explained Variance Ratio': explained_variance
})

print(pca_results)
print("\nTotal Variance Explained by All Components:", explained_variance.sum())
```

```python
# Step 1: Import necessary libraries
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Step 2: Load the Wine dataset
data = load_wine()
X = data.data
y = data.target

# Step 3: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Step 4: Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ---------------------------------------------------------
# Case 1: KNN on the original (scaled) dataset
# ---------------------------------------------------------
knn_original = KNeighborsClassifier(n_neighbors=5)
knn_original.fit(X_train_scaled, y_train)
y_pred_original = knn_original.predict(X_test_scaled)
accuracy_original = accuracy_score(y_test, y_pred_original)
print("Accuracy on original dataset:", accuracy_original)

# ---------------------------------------------------------
# Case 2: Apply PCA (retain top 2 principal components)
# ---------------------------------------------------------
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Train KNN on PCA-transformed data
knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_train_pca, y_train)
y_pred_pca = knn_pca.predict(X_test_pca)
accuracy_pca = accuracy_score(y_test, y_pred_pca)
print("Accuracy on PCA-transformed dataset (2 components):", accuracy_pca)
```

**Question 9: Train a KNN Classifier with different distance metrics (euclidean,
manhattan) on the scaled Wine dataset and compare the results.**
(Include your Python code and output in the code box below.)

Answer:

```python
# Step 1: Import necessary libraries
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Step 2: Load the Wine dataset
data = load_wine()
X = data.data
y = data.target

# Step 3: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Step 4: Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ---------------------------------------------------------
# Case 1: KNN with Euclidean distance (default)
# ---------------------------------------------------------
knn_euclidean = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn_euclidean.fit(X_train_scaled, y_train)
y_pred_euclidean = knn_euclidean.predict(X_test_scaled)
accuracy_euclidean = accuracy_score(y_test, y_pred_euclidean)
print("Accuracy with Euclidean distance:", accuracy_euclidean)

# ---------------------------------------------------------
# Case 2: KNN with Manhattan distance
# ---------------------------------------------------------
knn_manhattan = KNeighborsClassifier(n_neighbors=5, metric='manhattan')
knn_manhattan.fit(X_train_scaled, y_train)
y_pred_manhattan = knn_manhattan.predict(X_test_scaled)
accuracy_manhattan = accuracy_score(y_test, y_pred_manhattan)
print("Accuracy with Manhattan distance:", accuracy_manhattan)
```

**Question 10: You are working with a high-dimensional gene expression dataset to classify patients with different types of cancer. Due to the large number of features and a small number of samples, traditional models overfit.**
Explain how you would:
● Use PCA to reduce dimensionality
● Decide how many components to keep
● Use KNN for classification post-dimensionality reduction
● Evaluate the model
● Justify this pipeline to your stakeholders as a robust solution for real-world
biomedical data
(Include your Python code and output in the code box below.)

Answer:
In high-dimensional datasets such as gene expression data, where the number of features (genes) far exceeds the number of samples (patients), traditional models are prone to overfitting. This happens because the model learns noise rather than meaningful biological patterns. To address this issue, a combination of Principal Component Analysis (PCA) and K-Nearest Neighbors (KNN) can be used to build a robust and interpretable pipeline.

Step-by-Step Approach

1. Use PCA to Reduce Dimensionality
PCA transforms the original correlated gene features into a smaller set of uncorrelated principal components, which capture the majority of the data’s variance. This reduces noise and redundancy while preserving the most important biological signals.

2. Decide How Many Components to Keep
We select the number of principal components such that the cumulative explained variance ratio exceeds a threshold (e.g., 95%). This ensures minimal information loss while significantly reducing dimensionality.

3. Use KNN for Classification
After dimensionality reduction, a KNN classifier is trained on the transformed dataset. KNN is suitable here because it performs well in reduced, noise-free spaces and does not assume a specific data distribution.

4. Evaluate the Model
Model performance is evaluated using accuracy and cross-validation. Cross-validation ensures that the model generalizes well despite the limited sample size.

5. Justify the Pipeline to Stakeholders
This pipeline is robust for biomedical data because:

PCA mitigates overfitting by compressing high-dimensional features into a few informative components.

KNN provides transparent, interpretable predictions that clinicians can trust.

The combination ensures better generalization on unseen patient samples, making it suitable for diagnostic or prognostic modeling.

```python
# Step 1: Import required libraries
from sklearn.datasets import load_wine  # Simulating gene expression data for example
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Load dataset (using Wine dataset as a stand-in for gene expression data)
data = load_wine()
X = data.data
y = data.target

# Step 3: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Step 4: Standardize features (important for PCA and KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 5: Apply PCA - retain 95% of variance
pca = PCA(0.95)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Display number of components chosen
print("Number of PCA components retained:", pca.n_components_)

# Step 6: Train KNN classifier on PCA-transformed data
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_pca, y_train)
y_pred = knn.predict(X_test_pca)

# Step 7: Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy after PCA + KNN:", accuracy)

# Step 8: Cross-validation to check model stability
cv_scores = cross_val_score(knn, X_train_pca, y_train, cv=5)
print("Cross-Validation Accuracy (Mean):", np.mean(cv_scores))

# Step 9: Plot explained variance ratio
plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Principal Components')
plt.grid(True)
plt.show()
```
